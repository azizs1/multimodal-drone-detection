```python
#Install YOLO Libraries
!pip install ultralytics
from ultralytics import YOLO
```

```python
#Load Model (options include yolov8n.pt (smallest), yolov8s.pt, yolov8m.pt, yolov8l.pt, yolov8x.pt).
model1 = YOLO('yolov8n.pt')

#Copy data to local Colab storage for faster caching.
!cp -r "/content/drive/MyDrive/datasets/zenodo_visual_no_augmentation" /content/

#Train visual_no_augmentation model.
model1.train(
    data='/content/zenodo_visual_no_augmentation/data.yaml',
    epochs=50,
    project='/content/drive/MyDrive/runs',
    name='visual_no_augmentation_baseline'
)
```

```python
#Load Model (options include yolov8n.pt (smallest), yolov8s.pt, yolov8m.pt, yolov8l.pt, yolov8x.pt).
model2 = YOLO('yolov8n.pt')

#Copy data to local Colab storage for faster caching.
!cp -r "/content/drive/MyDrive/datasets/zenodo_thermal_no_augmentation" /content/

#Train thermal_no_augmentation model.
model2.train(
    data='/content/zenodo_thermal_no_augmentation/data.yaml',
    epochs=50,
    project='/content/drive/MyDrive/runs',
    name='thermal_no_augmentation_baseline'
)
```

```python
#NOTE: Takes about 3 hours to train, so I waited to train this at baseline after Sprint 1

#Load Model (options include yolov8n.pt (smallest), yolov8s.pt, yolov8m.pt, yolov8l.pt, yolov8x.pt).
model3 = YOLO('yolov8n.pt')

#Copy data to local Colab storage for faster caching.
!cp -r "/content/drive/MyDrive/datasets/zenodo_thermal_augmented" /content/

#Train thermal_augmented model.
model3.train(
    data='/content/zenodo_thermal_augmented/data.yaml',
    epochs=50,
    project='/content/drive/MyDrive/runs',
    name='thermal_augmented_baseline'
)
```

```python
#Load trained model and evaluate.
model1 = YOLO('/content/drive/MyDrive/runs/visual_no_augmentation_baseline/weights/best.pt')
model2 = YOLO('/content/drive/MyDrive/runs/thermal_no_augmentation_baseline/weights/best.pt')
#model3 = YOLO('/content/drive/MyDrive/runs/thermal_augmented_baseline/weights/best.pt')

#Validate on the test set.
metrics1 = model1.val(data='/content/zenodo_visual_no_augmentation/data.yaml', split='test', plots=True)
metrics2 = model2.val(data='/content/zenodo_thermal_no_augmentation/data.yaml', split='test', plots=True)
#metrics3 = model3.val(data='/content/zenodo_thermal_augmented/data.yaml', split='test', plots=True)
```

## Basic Metrics

It is clear that both of the initial non-augmented visual and thermal models performed very will, with mAP50 (precision) at 94.5% and 97.9% respectively. Both had low false positives and false negatives indicated by the P and R statistics above. For more information and visuals on this information, check out the runs folder for the corresponding model.
